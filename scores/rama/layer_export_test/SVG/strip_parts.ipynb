```python
from xml.dom import minidom
import json
```

```python

def _set_attributes(element, **attributes):
    for key, value in attributes.items():
        element.setAttribute(key, value)
        
def create_svg_object():
    imp = minidom.getDOMImplementation(features=[('xml', None)])
    doctype = imp.createDocumentType(
        'svg',
        '-//W3C//DTD SVG 1.1//EN',
        'http://www.w3.org/Graphics/SVG/1.1/DTD/svg11.dtd'
    )
    document = imp.createDocument(None, 'svg', doctype)
    _set_attributes(document.documentElement, version='1.1',
                    xmlns='http://www.w3.org/2000/svg', encoding="UTF-8")
    return document
```

```python
artboards = []
pages = []
for i in range(20):
    strI = str(i+1)
    if i < 9:
        strI = "0"+strI
    pages.append(strI)
    artboards.append( minidom.parse('echoic_orgArtBoards-'+strI+'.svg') )

npages = len(artboards)
```

```python
npages = len(artboards)
```

```python

svg = artboards[0].getElementsByTagName('svg')[0].cloneNode(1)
nodes = svg.childNodes

for n in nodes:
    tag = n.getAttribute('id')
    if tag.startswith('Layer'):
        
        newdoc = create_svg_object()
        root = newdoc.getElementsByTagName('svg')[0]
        f = open(tag+'.svg', 'w')

        for i in range(npages):
            a = artboards[i]
            page = pages[i]
            _svg = a.getElementsByTagName('svg')[0].cloneNode(1)
            _nodes = _svg.childNodes
            for _n in _nodes:
                _tag = _n.getAttribute('id')
                if _tag == tag: # or _tag == 'count':
                    _n.setAttribute('id', _tag +'-'+ page )
                    root.appendChild(_n)

        f.write( newdoc.toprettyxml(encoding='utf8') )
        f.close()
```
